```python
import pandas as pd
```

```python
ds = pd.read_csv("carrr.csv")
ds
```

```python
a = pd.get_dummies(ds.CarModel)
a
```

```python
merg = pd.concat([ds,a],axis="columns")
```

```python
merg
```

```python
f = merg.drop(['CarModel','Mercedez Benz C class'],axis='columns')
"""dummny variable trap"""
```

```python
f 
```

```python
f = f.drop(['Unnamed: 4'],axis = "columns")
```

```python
f
```

```python
from sklearn.linear_model import LinearRegression
mod = LinearRegression()
```

```python
x = f.drop(['Sell Price($)'],axis = "columns")
```

```python
x
```

```python
y = f.drop(["Mileage","Age(yrs)","Audi A5","BMW X5"],axis = "columns")
```

```python
y
```

```python
mod.fit(x,y)
```

```python
mod.predict([[35000,3,0,1]])
```

```python
mod.score(x,y)
```
