```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
m = 0
b = 0
it = 100000
ll = 0.01 
n = len(x)
plt.scatter(x,y)
a = range(1,6)
for i in range(it):
    yp = m*x+b
    dm = -(2/n)*sum(x*(y-yp))
    db = -(2/n)*sum(y-yp)
    m = m -(ll*dm)
    b = b -(ll*db)
    print("m = {},b={},i={}".format(m,b,i))
    g = m*a + b
    plt.plot(a,g) 







```
